This split is very close to the requested 2/3 (≈66.6%) for training and 1/3 (≈33.3%) for testing. The tiny difference (0.1%) comes from rounding when the total number of rows (392) is not perfectly divisible by 3. This is completely normal and acceptable.

```python
# (a) Fit a logistic regression model that uses `income` and `balance` to predict `default`
data = pd.read_csv('Default.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data['default'] = data['default'].replace({'Yes': 1, 'No': 0})
data['default'] = data['default'].astype(int)
```

```python
data.dtypes
```

```python
#(a)Fit a logistic regression model that uses `income` and `balance` to predict `default`.

#Setting a random seed for reproducibility
np.random.seed(123)


X = data[['income', 'balance']]          
X = sm.add_constant(X)
y = data['default']
modele_logit = sm.Logit(y, X).fit()

# Display the full summary (coefficients, standard errors, p-values, etc.)
print(modele_logit.summary())

# Show only the coefficients
print("\n=== Coefficients ===")
print(modele_logit.params)

# Display odds ratios (exponentiated coefficients) — very useful for interpretation
print("\n=== Odds Ratios (exp(coefficients)) ===")
print(np.exp(modele_logit.params))
```

```python
# (b) Using the validation set approach, estimate the test error of this model. In order to do this, you must perform the following steps:
# b.i. Split the sample set into a training set and a validation set

from sklearn.model_selection import train_test_split

# We split the data into 50% training and 50% validation 
train_df, val_df = train_test_split(data, test_size=0.5, random_state=123, stratify=data['default'])

# Display the sizes to confirm
print(f"Total observations:          {len(data)}")
print(f"Training set size:           {len(train_df)} ({100*len(train_df)/len(data):.1f}%)")
print(f"Validation (test) set size:  {len(val_df)} ({100*len(val_df)/len(data):.1f}%)")

# Quick check that the proportion of defaults is similar in both sets (thanks to stratify)
print("\nProportion of defaults:")
print(f"Overall:     {data['default'].mean():.4f}")
print(f"Training:    {train_df['default'].mean():.4f}")
print(f"Validation:  {val_df['default'].mean():.4f}")
```

```python
# b.ii. Fit the logistic regression on the TRAINING SET only

# Prepare features and target for the training set
X_train = train_df[['income', 'balance']]     
X_train = sm.add_constant(X_train)           
y_train = train_df['default']                  

# Fit the model
logit_train = sm.Logit(y_train, X_train).fit(disp=0)  # disp=0 suppresses convergence messages

# Display results
print(logit_train.summary())

# Show coefficients and odds ratios
print("\n=== Coefficients (training model) ===")
print(logit_train.params)

print("\n=== Odds Ratios ===")
print(np.exp(logit_train.params))
```

```python
# b.iii. Predict on the VALIDATION set

# Prepare the validation features 
X_val = val_df[['income', 'balance']]
X_val = sm.add_constant(X_val)        

# 1. Predicted probabilities of default 
val_pred_prob = logit_train.predict(X_val)

# 2. Predicted class: 1 if probability > 0.5, else 0
val_pred_class = (val_pred_prob >= 0.5).astype(int)

# Show first 10 results as example
print("First 10 validation observations:")
print(pd.DataFrame({
    'True default': val_df['default'].head(10).values,
    'Pred prob': val_pred_prob.head(10).round(4),
    'Pred class (threshold 0.5)': val_pred_class.head(10)
}))
```

```python
# b.iv. Calculate the validation set error

# True labels from validation set
y_val_true = val_df['default']

# Predicted classes from step iii (threshold = 0.5)
y_val_pred = val_pred_class

# Number of misclassified observations
misclassified = (y_val_true != y_val_pred).sum()

# Misclassification error rate
validation_error = misclassified / len(y_val_true)

# Display results
print(f"Validation set size       : {len(y_val_true)}")
print(f"Misclassified observations : {misclassified}")
print(f"Validation set error rate : {validation_error:.4f}  ({validation_error*100:.2f}%)")
```

```python
# b. Write a function, `boot_fn()`, that takes as input the `Default` data set as well as an index of the observations, and that outputs the coefficient
# estimates for `income` and `balance` in the multiple logistic regression model.

X_full = sm.add_constant(data[['income', 'balance']])
y_full = data['default']

def boot_fn(data = data, index=None):

    if index is None:
        index = np.arange(len(data))  
    
    # Create bootstrap sample
    X_boot = X_full.iloc[index]      
    y_boot = y_full.iloc[index]
    
    # Fit logistic regression on the bootstrap sample
    model_boot = sm.Logit(y_boot, X_boot).fit(disp=0)  
    
    # Extract coefficients
    coef_const   = model_boot.params['const']
    coef_income  = model_boot.params['income']
    coef_balance = model_boot.params['balance']
    
    # Return as a clean dictionary (easy to collect in bootstrap loop)
    return {
        'const'  : coef_const,
        'income' : coef_income,
        'balance': coef_balance
    }

# Quick test of the function on the original data
np.random.seed(1)
test_result = boot_fn()
print("Test on original data:")
print(test_result)
```

```python
#c. Use the `boot()` function together with your `boot.fn()` function to estimate the standard errors of the logistic regression coefficients for
#income and balance.

# Set seed for perfect reproducibility
np.random.seed(123)

# --- The boot_fn() function from part (b) ---
def boot_fn(index):
    """Fit logistic regression on a bootstrap sample and return coefficients"""
    X_boot = X_full.iloc[index]
    y_boot = y_full.iloc[index]
    model = sm.Logit(y_boot, X_boot).fit(disp=0)
    return {
        'const'   : model.params['const'],
        'income'  : model.params['income'],
        'balance' : model.params['balance']
    }

# --- Run the bootstrap (R = 5000 replications – more stable than 1000) ---
R = 5000
n = len(data)

bootstrap_results = []

for i in range(R):
    boot_indices = np.random.choice(n, size=n, replace=True)
    coefs = boot_fn(boot_indices)
    bootstrap_results.append(coefs)

# Convert to DataFrame for easy computation
boot_df = pd.DataFrame(bootstrap_results)

# Compute bootstrap standard errors
boot_se = boot_df.std()

# Original standard errors from the usual formula (part a)
original_model = sm.Logit(y_full, X_full).fit(disp=0)
original_se = original_model.bse

# --- Display results ---
print("="*65)
print("COMPARISON : STANDARD ERRORS (Logistic Regression)")
print("="*65)
comparison = pd.DataFrame({
    'Standard formula (glm)' : original_se,
    'Bootstrap (R=5000)'     : boot_se
})
print(comparison.round(8))

print("\nFinal bootstrap standard errors to report:")
print(f"income  : {boot_se['income']:.10f}")
print(f"balance : {boot_se['balance']:.8f}")
print("="*65)
```
